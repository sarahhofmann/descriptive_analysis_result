```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getNumber(data, digit):
    data_number = pd.DataFrame(data[data[0] == digit])
    cleaned_data = data_number.iloc[:,1:]
    return cleaned_data
```

```python
def getNumberImage(data, digit):
    cleaned_data = getNumber(data, digit)
    df1 = pd.DataFrame(cleaned_data.mean())
    return df1.values.reshape(16,16)
```
